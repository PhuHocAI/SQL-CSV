```python
import pyodbc
import csv

# Thông tin kết nối đến SQL Server
server = 'Name of server'  # Thay thế bằng tên server của bạn
database = 'Name of database'  # Thay thế bằng tên database của bạn

# Danh sách các bảng cần xuất
tables_to_export = ["Name of columms"]

# Chuỗi kết nối
conn_str = (
    r'DRIVER={SQL Server};'
    r'SERVER=' + server + ';'
    r'DATABASE=' + database + ';'
    r'Trusted_Connection=yes;'  # Thêm tùy chọn này
)

try:
    # Kết nối đến SQL Server
    cnxn = pyodbc.connect(conn_str)
    cursor = cnxn.cursor()

    for table_name in tables_to_export:
        # Tạo câu truy vấn SQL
        query = f"SELECT * FROM dbo.{table_name}"

        # Thực thi truy vấn
        cursor.execute(query)

        # Lấy tên cột làm header cho file CSV
        column_names = [column[0] for column in cursor.description]

        # Tạo tên file CSV
        csv_file_name = f"{table_name}.csv"

        # Mở file CSV để ghi
        with open(csv_file_name, 'w', newline='', encoding='utf-8') as csvfile:
            writer = csv.writer(csvfile)

            # Ghi header
            writer.writerow(column_names)

            # Ghi dữ liệu
            for row in cursor:
                writer.writerow(row)

        print(f"Đã xuất bảng {table_name} thành file {csv_file_name}")

except pyodbc.Error as ex:
    sqlstate = ex.args[0]
    print(f"Lỗi kết nối hoặc truy vấn SQL: {sqlstate}")

finally:
    # Đóng kết nối
    if cnxn:
        cnxn.close()
```
